```python
emprestimos = (
    df.groupby("borrower_country")["commitment_USD"]
    .sum()
    .sort_values(ascending=False) / 1e9
)

emprestimos.head(20)
```

```python
top10 = (
    df.groupby("borrower_country")["commitment_USD"]
    .sum()
    .sort_values(ascending=False)
    .head(10) / 1e9
)

top10.plot(kind="bar")

plt.title("Top 10 países que mais receberam empréstimos chineses")
plt.ylabel("Bilhões de USD")
plt.xlabel("País")
plt.show()
```

```python
df["year"] = df["contract_id"].str.extract(r"_(\d{4})_").astype(int)
ultimo_ano = df["year"].max()
ultimosanos = df[df["year"] >= ultimo_ano - 13]
ultimosanos.groupby("borrower_country")["commitment_USD"] \
         .sum() \
         .sort_values(ascending=False)/1e9


top10 = (
  ultimosanos.groupby("borrower_country")["commitment_USD"]
    .sum()
    .sort_values(ascending=False)
    .head(10) / 1e9
)

top10.plot(kind="bar")

plt.title("Top 10 países que mais receberam empréstimos chineses nos últimos 13 anos")
plt.ylabel("Bilhões de USD")
plt.xlabel("País")
plt.show()

```

```python
df["year"] = df["contract_id"].str.extract(r"_(\d{4})_").astype(int)
ultimo_ano = df["year"].max()
ultimosanos = df[df["year"] >= ultimo_ano - 13]


region = (
  ultimosanos.groupby("borrower_region")["commitment_USD"]
    .sum()
    .sort_values(ascending=False)/ 1e9
)

region.plot(kind="bar")

plt.title("As regiões que mais receberam empréstimos chineses nos últimos 13 anos")
plt.ylabel("Bilhões de USD")
plt.xlabel("Região")
plt.show()
```

```python
latam = df[
    (df["borrower_region"] == "Latin America & Caribbean") &
    (df["year"] >= ultimo_ano - 13)
]

ranking = (
    latam.groupby("borrower_country")["commitment_USD"]
    .sum()
    .sort_values(ascending=False) / 1e9
)



ranking.plot(kind="bar")

plt.title("Países da América Latina que mais receberam empréstimos chineses (últimos 13 anos)")
plt.ylabel("Bilhões de USD")
plt.xlabel("País")
plt.show()
```
